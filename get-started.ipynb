### Track every experiment and visualize with MLflow
Start MLflow tracking server locally
```
mlflow ui
```

````python
import mlflow
````

````python
mlflow.set_tracking_uri("http://127.0.0.1:5000")
````

````python
mlflow.set_experiment("setup local experiment")
with mlflow.start_run():
    mlflow.log_metric("accuracy", 0.95)
    mlflow.log_metric("version", 1.0)
````
